```python
'''
For a 3 by 3 matrix:
| a b c |
| d e f |
| g h i |

The inverse is given by:
1 / ((a * e * i) + (b * f * g) + (c * d * h) - (a * f * h) - (b * d * i) - (c * e * g)) * | (e * i - f * h) (c * h - b * i) (b * f - c * e) |
| (f * g - d * i) (a * i - c * g) (c * d - a * f) |
| (d * h - e * g) (b * g - a * h) (a * e - b * d) |
'''

import numpy as np

# Set the global configuration for float display
np.set_printoptions(precision=4)

# matrix = np.array([[2, 3, 1],
# [4, 1, 7],
# [0, 5, 2]])

matrix = np.array([[17, 17, 5],
                   [21, 18, 21],
                   [2, 2, 19]])

a = matrix[0][0]
b = matrix[0][1]
c = matrix[0][2]

d = matrix[1][0]
e = matrix[1][1]
f = matrix[1][2]

g = matrix[2][0]
h = matrix[2][1]
i = matrix[2][2]

# Calculate the determinant
det = (a * e * i + b * f * g + c * d * h) - (a * f * h + b * d * i + c * e * g)

print("Matrix is:\n")
for row in matrix:
    print(row)

print("Determinant: ", det)

# Check if the determinant is zero (matrix is singular)
if det == 0:
    print("The matrix is singular, and its inverse does not exist.")
else:
    # Calculate the elements of the inverse matrix
    inv_a = (e * i - f * h) / det
    inv_b = (c * h - b * i) / det
    inv_c = (b * f - c * e) / det
    inv_d = (f * g - d * i) / det
    inv_e = (a * i - c * g) / det
    inv_f = (c * d - a * f) / det
    inv_g = (d * h - e * g) / det
    inv_h = (b * g - a * h) / det
    inv_i = (a * e - b * d) / det

    # Create the inverse matrix
    inverse = [[inv_a, inv_b, inv_c],
               [inv_d, inv_e, inv_f],
               [inv_g, inv_h, inv_i]]

    inverse_mod_26 = np.array(inverse) % 26

    # Print the inverse matrix
    for row in inverse_mod_26:
        print(row)
```
